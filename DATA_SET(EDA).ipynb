```
import pandas as pd
```

```
corona_confirmed_df=pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/archived_data/archived_time_series/time_series_2019-ncov-Confirmed.csv")
```

```
corona_confirmed_df.head()
```

```
corona_deaths_df=pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/archived_data/archived_time_series/time_series_2019-ncov-Deaths.csv")
```

```
corona_deaths_df.head()
```

```
corona_recovered_df=pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/archived_data/archived_time_series/time_series_2019-ncov-Recovered.csv")
```

```
corona_recovered_df.head()
```

```
corona_recovered_df.fillna(0,inplace=True)
```

```
corona_recovered_df.head()
```

```
corona_recovered_df=corona_recovered_df.melt(id_vars=["Province/State","Country/Region","Lat","Long"])
```

```
corona_recovered_df=corona_recovered_df.rename({"variable":"Date","value":"recovered"},axis="columns")
```

```
corona_recovered_df.head()
```

```
corona_deaths_df.fillna(0,inplace=True)
```

```
corona_deaths_df.head()
```

```
corona_deaths_df=corona_deaths_df.melt(id_vars=["Province/State","Country/Region","Lat","Long"])
```

```
corona_deaths_df.head()
```

```
corona_deaths_df=corona_deaths_df.rename({"variable":"Date","value":"deaths"},axis="columns")
```

```
corona_deaths_df.head()
```

```
corona_confirmed_df.fillna(0,inplace=True)
```

```
corona_confirmed_df=corona_confirmed_df.melt(id_vars=["Province/State","Country/Region","Lat","Long"])
```

```
corona_confirmed_df.head()
```

```
corona_confirmed_df=corona_confirmed_df.rename({"variable":"Date","value":"confirmed"},axis="columns")
```

```
corona_confirmed_df.head()
```

```
combined_df=[corona_confirmed_df,corona_deaths_df,corona_recovered_df]
```

```
combined_df=[df.set_index(['Province/State',"Country/Region","Lat","Long","Date"]) for df in combined_df]
```

```
combined_df=combined_df[0].join(combined_df[1:])
```

```
combined_df.head()
```

```
combined_df=combined_df.reset_index()
```

```
combined_df.head()
```

```
combined_df[["Lat","Long","confirmed","deaths","recovered"]]=combined_df[["Lat","Long","confirmed","deaths","recovered"]].apply(pd.to_numeric)
```

```
combined_df.dtypes
```

```
combined_df[["Date"]]=combined_df[["Date"]].apply(pd.to_datetime)
```

```
combined_df.dtypes
```

```
combined_df=combined_df.rename({"Province/State":"State","Country/Region":"Country"},axis="columns")
```

```
combined_df.head()
```

```
combined_df.isnull().any()
```

```
combined_df.count()
```

```
combined_df[combined_df.Country=="US"]
```

```
US_states=pd.read_csv("https://raw.githubusercontent.com/jasonong/List-of-US-States/master/states.csv")
```

```
US_states.head()
```

```
US_states=US_states.set_index("Abbreviation")
```

```
US_states.head()
```

```
US_states.loc["AZ"][0]
```

```
combined_df["City"]=combined_df.apply(lambda x: x["State"].split(",")[0] if(x["Country"]=="US") else " ",axis=1)
```

```
combined_df["City"]
```

```
def convertcase(State):
  try:
    output=State.split(",")[1].strip()
    output=US_states.loc[output.upper()][1]
  except: 
    output=State
  return output
```

```
combined_df["state Cleaned"]=combined_df.apply(lambda x: convertcase(x["State"]) if(x["Country"]=="US") else x["State"],axis=1)
```
